```python
plt.hist(data['Cost'], color = 'blue', edgecolor = 'black',
         bins = int(15))
plt.title('Logistic Cost')
plt.ylabel('Frequency')
plt.xlabel('Range')
```

```python
sns.distplot(data['Cost'], hist=True, kde=True, 
             bins=int(15), color = 'darkblue', 
             hist_kws={'edgecolor':'black'},
             kde_kws={'linewidth': 4})
```
